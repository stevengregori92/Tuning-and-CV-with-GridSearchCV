```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('titanic.csv', index_col = 'PassengerId')
data
```

```python
data.drop(columns = ['Name', 'Ticket', 'Cabin', 'Age'], inplace = True)
```

```python
X = data.drop(columns=['Survived'])
y = data.Survived

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state = 42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

#Train Val Test Split

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
```

```python
numerical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy = 'mean')),
    ('scaler', MinMaxScaler())
])

categorical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy = 'most_frequent')),
    ('scaler', OneHotEncoder())
])
```

```python
from sklearn.compose import ColumnTransformer
```

```python
X_train.head()
```

```python
preprocessor = ColumnTransformer([
    ('numeric', numerical_pipeline, ['Fare', 'SibSp', 'Parch']),
    ('categoric', categorical_pipeline, ['Pclass', 'Sex', 'Embarked'])
])
```

#PIPELINE

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
pipeline = Pipeline([
    ('prep', preprocessor),
    ('algo', KNeighborsClassifier())
])
```

```python
pipeline.fit(X_train, y_train)
```

```python
pipeline.score(X_test, y_test)
```

#GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
pipeline.get_params()
```

```python
parameters = {
    'algo__n_neighbors' : range(1, 50,2),
    'algo__weights' : ['uniform', 'distance'],
    'algo__p' : [1,2]
}

model = GridSearchCV(pipeline, parameters, cv=3, n_jobs=-1, verbose = 1)
model.fit(X_train, y_train)
```

```python
model.cv_results_
```

```python
pd.DataFrame(model.cv_results_).sort_values('rank_test_score', ascending = True)
```

```python
model.best_params_
```

```python
model.score(X_train, y_train), model.score(X_test, y_test)
```
